# CO DATA

```python
import rasterio
import numpy as np
from matplotlib import pyplot
import matplotlib.pyplot as plt
from rasterio import plot
%matplotlib inline
nir = rasterio.open("20030707/LT05_CU_006010_20030707_20190106_C01_V01_SRB4.tif")
swir = rasterio.open("20030707/LT05_CU_006010_20030707_20190106_C01_V01_SRB7.tif")

# each .tif file is a single band
nir.count
```

```python
from rasterio.plot import show
nir = nir.read()
swir = swir.read()
fig, (axir, axswir) = pyplot.subplots(1,2, figsize=(14,7))

axir = show(nir, ax=axir, cmap= "Greys", title="IR channel")
axswir = show(swir, ax=axswir, cmap= "Greys_r", title="SWIR channel")
```

```python
#ndvi - not effective at marking burn area
#faint small specks in top right corner is vegetation from farmland
red = rasterio.open("20030707/LT05_CU_006010_20030707_20190106_C01_V01_SRB3.tif")
nir = rasterio.open("20030707/LT05_CU_006010_20030707_20190106_C01_V01_SRB4.tif")

red = red.read()
nir = nir.read()
red = red.astype(float)
nir = nir.astype(float)
np.seterr(divide='ignore', invalid='ignore')
ndvi = ((nir - red)/(nir + red))
plot.show(ndvi, cmap = "Reds")
```

```python
#Prefire - nbr
nir = rasterio.open("20030605/LT05_CU_006010_20030605_20190109_C01_V01_SRB4.tif")
swir = rasterio.open("20030605/LT05_CU_006010_20030605_20190109_C01_V01_SRB7.tif")

nir = nir.read()
swir = swir.read()
nir = nir.astype(float)
swir = swir.astype(float)
np.seterr(divide='ignore', invalid='ignore')
nbr_prefire = ((nir - swir) / (nir + swir))
plot.show(nbr_prefire, cmap = "Greys")
```

```python
#Postfire - nbr
nir = rasterio.open("20030707/LT05_CU_006010_20030707_20190106_C01_V01_SRB4.tif")
swir = rasterio.open("20030707/LT05_CU_006010_20030707_20190106_C01_V01_SRB7.tif")

nir = nir.read()
swir = swir.read()
nir = nir.astype(float)
swir = swir.astype(float)
np.seterr(divide='ignore', invalid='ignore')
nbr_postfire = ((nir - swir) / (nir + swir))
plot.show(nbr_postfire, cmap = "Greys")
```

```python
# difference in NBR shows relative burn severity
#Pre-fireNBR - Post-fireNBR = dNBR
dnbr = nbr_prefire - nbr_postfire
plot.show(dnbr, cmap = "RdYlGn_r")
```
